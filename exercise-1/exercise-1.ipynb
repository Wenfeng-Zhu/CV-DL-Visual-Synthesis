3. Plot a random crop of size 256  256 from the image.

```python
def random_crop(img, x, y):
    img_crop = img[y:(y + 256), x:(x + 256)]
    plt.imshow(img_crop)
    pltTitle = 'Random Crop of 256 X 256 X 3 starts from (' + str(x) + ' , ' + str(y) + ')'
    plt.title(pltTitle)
    plt.show()
    return img_crop


start_x, start_y = random.randint(0, (width - 257)), random.randint(0, (height - 257))
img_crop = random_crop(img, start_x, start_y)
```

4. Convert the cropped out image patch to grayscale and plot

```python
def grayscale(img):
    R, G, B = img[:, :, 0], img[:, :, 1], img[:, :, 2]
    imgGray = 0.2989 * R + 0.5870 * G + 0.1140 * B
    plt.imshow(imgGray, cmap='gray')
    pltTitle = 'Grayscale images of Random Crop'
    plt.title(pltTitle)
    plt.show()
    return imgGray


img_crop_grayscale = grayscale(img_crop)
```

5. Insert the grayscale patch back into the original image (note that the grayscale patch only has one channel) and plot.

```python
def insert_grayscale(img, img_crop_grayscale, x, y):
    img[y:(y + 256), x:(x + 256)] = np.stack((img_crop_grayscale,) * 3, axis=-1)
    plt.imshow(img)
    pltTitle = 'Insert the grayscale patch back into the original image'
    plt.title(pltTitle)
    plt.show()
    plt.clf()
    return img

img_insert_grayscale = insert_grayscale(img, img_crop_grayscale, start_x, start_y)
```

6. Resize the image with the inserted grayscale patch with a factor of 1/2 for both height and width.

```python
def resize(img, factor):
    new_size = int(img.shape[1] * factor), int(img.shape[0] * factor)
    img_pil = Image.fromarray(img)
    img_resize = img_pil.resize(new_size)
    plt.imshow(img_resize)
    pltTitle = 'Resize image with 1/2 of grayscale-inserted imag'
    plt.title(pltTitle)
    plt.show()

    
resize(img_insert_grayscale, 0.5)
```

2. Implement a function that computes a Gaussian filter as defined in Eq. 2. The function should take in the kernel size, mean, and variance as inputs. Then it should output the kernel represented as an nn NumPy array. Plot the filter as a grayscale image of size 200  200.

```python
def gaussian_kernel(kernel_size, mean, variance):
    kernel = np.zeros((kernel_size, kernel_size))
    for i in range(kernel_size):
        for j in range(kernel_size):
            x, y = i - mean[0], j - mean[1]
            # gaussian kernel function
            kernel[i, j] = np.exp(-(x ** 2 + y ** 2) / (2 * variance)) / (2 * np.pi * variance)
    # kernel = normalized_grayscale(kernel)
    print('Gaussian Kernel of size', kernel_size, "x", kernel_size, ", mean", mean, ", variance", variance)
    return kernel


gaussian_filter = gaussian_kernel(200, [80, 100], 50)
plt.imshow(gaussian_filter, cmap='gray')
plt.title('Gaussian Kernel')
plt.show()
```

```python
def convolution_operation(imgGray, conv_filter):
    height, width = imgGray.shape
    size = len(conv_filter)
    # create a new numpy array with padding to make sure the filtered image has the same size as the input image
    # It must be noted, however, that filters of odd size will facilitate the calculation
    img_conv = []
    for i in range(height-size):
        row = []
        for j in range(width-size):
            input = imgGray[i:i + len(conv_filter), j:j + len(conv_filter)]
            row.append(np.sum(np.multiply(input, conv_filter)))
        img_conv.append(row)
    return np.array(img_conv)


imgGray = grayscale(img)
img_conv_gaussian = convolution_operation(imgGray, gaussian_kernel(7, [4, 3], 30))
plt.imshow(img_conv_gaussian, cmap="gray")
plt.title('Blurred image')
plt.show()
```

4. There are many other useful filters that are utilized to analyze images e.g. edge detectors. Implement a Laplace filter (also called as discrete Laplace operator) and apply it to the image dog.jpg (in grayscale) using your convolution operation. How does it detect the edges? Think about the mathematical operation Laplacian filter corresponds to.

```python
def laplacian_of_gaussian(x, y, sigma):
    # Formatted this way for readability
    nom = (y ** 2) + (x ** 2) - 2 * (sigma ** 2)
    denom = 2 * np.pi * (sigma ** 6)
    expo = np.exp(-((x ** 2) + (y ** 2)) / (2 * (sigma ** 2)))
    return (nom * expo / denom)


def create_log(sigma, kernel_size):
    w = (np.ceil(float(kernel_size) * float(sigma)))
    if w % 2 == 0:
        w = w + 1
    l_o_g_mask = []
    w_range = int(np.floor(w / 2))
    for i in range(-w_range, w_range + 1):
        for j in range(-w_range, w_range + 1):
            l_o_g_mask.append(laplacian_of_gaussian(i, j, sigma))
    l_o_g_mask = np.array(l_o_g_mask)
    l_o_g_mask = l_o_g_mask.reshape(int(w), int(w))
    return l_o_g_mask


def laplacian_filter(imgGray, kernel_size, sigma):
    log_mask = create_log(sigma, kernel_size)
    img_log = convolution_operation(imgGray, log_mask)
    norm = (img_log - np.min(img_log)) / (np.max(img_log) - np.min(img_log))
    img_log = img_log * norm * 255
    return img_log


img_laplacian = laplacian_filter(imgGray, 7, 1)
plt.imshow(img_laplacian, cmap="gray")
plt.title('Laplace operation')
plt.show()
```
